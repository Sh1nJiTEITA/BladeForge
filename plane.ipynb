```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
```

```python
class plane:
    def __init__(self, A,B,C,D):
        self.A = A
        self.B = B
        self.C = C
        self.D = D
```

```python
def get_plane_coo(p:plane,x,y):
    return -(p.A * x + p.B * y + p.D)/(p.C)
    

class rectangle:
    def __init__(self, x,y,side):
        self.x = x
        self.y = y
        self.side = side
    
    def get_2D_vertices(self):
        x = [self.x, self.x, self.x + self.side, self.x + self.side]
        y = [self.y, self.y + self.side, self.y + self.side, self.y]
        z = [0, 0, 0, 0]
        return x, y, z
    
    def get_3D_vertices(self):
        x = [self.x, self.x, self.x + self.side, self.x + self.side]
        y = [self.y, self.y + self.side, self.y + self.side, self.y]
        z = [get_plane_coo(x[0], y[0]), get_plane_coo(x[1], y[1]), get_plane_coo(x[2], y[2]), get_plane_coo(x[3], y[3])]
        return x, y, z
```

```python
p = plane(5.0, 2.0, 3.0, -10.0)
xv = np.arange(0, 10.1, 0.1)
yv = np.arange(0, 10.1, 0.1)
x, y = np.array(list(product(xv, yv))).T
z = -(p.A * x + p.B * y + p.D) / p.C

rec = rectangle(0, 0, 2)
x_rec, y_rec, z_rec = rec.get_2D_vertices()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(x, y, z, cmap='viridis')
ax.plot_trisurf(x_rec, y_rec, z_rec)
```
